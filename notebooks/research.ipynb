## By looking at the first results of analysis things i should dive into are:

- In the KRONIKHASTALIK column I should find a way to categorize the patients I am assuming directly doing encoding would result in too many unique patient which would be hard for modelling phase. also same case for TANILAR and TEDAVIADI columns.


```python
df["UYGULAMASURESI"] = df["UYGULAMASURESI"].str.replace("Dakika", "", regex=False)
df["UYGULAMASURESI"] = df["UYGULAMASURESI"].str.strip()
df["UYGULAMASURESI"] = df["UYGULAMASURESI"].astype(int)
```

```python
df["UYGULAMASURESI_SAAT"] = df["UYGULAMASURESI"] / 60 
df["UYGULAMASURESI_SAAT"]
```

```python
df.rename(columns={"UYGULAMASURESI": "UYGULAMASURESI_DAKIKA"}, inplace=True)
```

```python
df["TEDAVISURESI"] = df["TEDAVISURESI"].str.replace("Seans", "", regex=False)
df["TEDAVISURESI"] = df["TEDAVISURESI"].str.strip()
df["TEDAVISURESI"] = df["TEDAVISURESI"].astype(int)
```

```python
df.rename(columns={"TEDAVISURESI": "TEDAVISURESI_SEANS_SAYI"}, inplace=True) # TEDAVISURESI_SEANS_SAYI will be predicted 
```

```python
df["TEDAVISURESI_TOPLAM_DAKIKA"] = df["TEDAVISURESI_SEANS_SAYI"] * df["UYGULAMASURESI_DAKIKA"]
df["TEDAVISURESI_TOPLAM_SAAT"]   = df["TEDAVISURESI_SEANS_SAYI"] * df["UYGULAMASURESI_SAAT"]
```

```python
for col in ["KRONIKHASTALIK", "BOLUM", "ALERJI", "TANILAR", "UYGULAMAYERLERI"]:
    unique_elems = set([item.strip() for sublist in df[col].dropna().str.split(",") for item in sublist])
    print(f"Unique Elems in col {col}: ", unique_elems)
    print(f"Len of unique elems in col {col}: ", len(unique_elems))
```

```python
# By doing reasearch I find out Hiportiroidizm, Hipotirodizm is a typo of Hipotiroidizm
# Same case for ALERJI column Volteren, Voltaren, VOLTAREN => Voltaren

df["KRONIKHASTALIK"] = df["KRONIKHASTALIK"].str.replace("Hiportiroidizm", "Hipotiroidizm", regex=False)
df["KRONIKHASTALIK"] = df["KRONIKHASTALIK"].str.replace("Hipotirodizm", "Hipotiroidizm", regex=False)

df["ALERJI"] = df["ALERJI"].str.replace("Volteren", "Voltaren", regex=False)
df["ALERJI"] = df["ALERJI"].str.replace("Voltaren", "Voltaren", regex=False)
df["ALERJI"] = df["ALERJI"].str.replace("VOLTAREN", "Voltaren", regex=False)
```

```python
d = {}
d["mean"] = []
d["mode"] = []

for col in df.columns:
    if np.issubdtype(df[col].dtype, np.number):
        d["mean"].append(col)
    else:
        d["mode"].append(col)
```

```python
agg_dict = {col: "mean" for col in d["mean"]}
agg_dict.update({col: lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan for col in d["mode"]})

agg_dict_fixed = agg_dict.copy()
agg_dict_fixed.pop("HASTANO")
grouped_by_hastano = df.groupby("HASTANO").agg(agg_dict_fixed).reset_index()
```
